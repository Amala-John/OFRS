```python
import numpy as np

# Define the specific attributes for which you want to create boxplots
attributes = ['Temperature']

# Create a DataFrame to store the cleaned data without outliers
data1= data.copy()

# Iterate over the specified attributes
for col in attributes:
    # Calculate the Z-scores for the current attribute
    z_scores = (data[col] - data[col].mean()) / data[col].std()
    
    # Define a threshold for Z-score (e.g., 2)
    z_score_threshold = 2
    
    # Identify the rows with Z-scores greater than the threshold
    outliers_idx = np.abs(z_scores) > z_score_threshold
    
    # Remove the outliers from the current attribute
    data1 = data1[~outliers_idx]
    
    # Create a boxplot for the cleaned attribute without outliers
    plt.figure(figsize=(5, 5))
    plt.boxplot(data1[col])
    plt.title(col)

# Display all the boxplots together
plt.tight_layout()
plt.show
```

```python
X = df.drop(columns=['DistFireNumber','Incident_Ward','Est_Persons_Displaced','Estimated_Dollar_Loss','month_of_report','Team_Assigned'])
```

```python
Y = df['month_of_report']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Create an instance of the Linear Regression model
lr_model = LinearRegression()

# Fit the model to the training data
lr_model.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = lr_model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(Y_test, Y_pred)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# Instantiate a RandomForestRegressor object
reg = RandomForestRegressor(max_depth=2)

# Fit the model to the training data
reg.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = reg.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(Y_test, Y_pred)
print("R-squared:", r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# Instantiate a DecisionTreeRegressor object
dt_regressor = DecisionTreeRegressor(max_depth = 2)

# Fit the model to the training data
dt_regressor.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = dt_regressor.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(Y_test, Y_pred)
print("R-squared:", r2)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# Instantiate a DecisionTreeRegressor object
lassoReg = linear_model.Lasso(alpha = 0.1)

# Fit the model to the training data
lassoReg.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = lassoReg.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(Y_test, Y_pred)
print("R-squared:", r2)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from tensorflow import keras

# Load and preprocess the dataset
# Assuming X_train and y_train are your input features and target variable respectively

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the neural network architecture
model = keras.models.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(num_features,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)  # For regression, no activation function in the final layer
])

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train_scaled, Y_train, epochs=10, batch_size=32, validation_data=(X_test_scaled, Y_test))

# Make predictions
y_pred = model.predict(X_test_scaled)

# Calculate mean squared error
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:", mse)
```
